# Verificação de Outliers

```python
# Visualizando distribuições com histogramas para ter intuição quanto a outliers

for col in df:

    sns.histplot(data=df, x=col, kde=True).set_title(f"Distribuição da variável {col}")
    plt.show()
```

A seguir, iremos utilizar o **Método do Quartil**

- Para cada coluna do dataset, desconsiderando a qualidade
    - Calcula-se o IQR
    - Realiza-se filtragem que identifica o IQR usando o critério da distância inter-quartil, para verificar os índices dos outliers
    - Exibe as colunas que possuem outlier e respectivas posições(índices)
        - Caso não haja outliers, uma mensagem indicando isto é impressa na tela. 

```python
## Método do Quartil.

for col in df.drop(columns='quality'):
    
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    aux_outliers = df[(df[col] < Q1-(IQR*1.5)) 
                   | (df[col] > Q3+(IQR*1.5))]
    
    indices_outliers = aux_outliers.index.tolist()
    
    if len(indices_outliers) >= 1:
        
        print(f"A coluna {col} tem {len(indices_outliers)} outliers!")
        print("\nOs índices deles são:\n")
        print(indices_outliers)

    else:
        
        print(f"Acoluna {col} não tem outliers!")
        
    print()
    print("="*80)
    print()
```

6) Calculo / visualização do intervalo de confiança de 90% para a média de cada uma das variáveis físico-químicas, agrupadas pelos níveis categóricos da variável resposta quality.

```python
# Geração do(s) gráfico(s)

for col in df.drop(columns='quality'):
    
    sns.barplot(data=df, x='quality', y=col, errorbar=('ci', 90), hue='quality')
    plt.show()
```

- Percebe-se que como existe uma diferença grande entre os vinhos da ponta, talvez não seja possível uma identificação de padrão adequada para os casos mais extremos, snendo interessante separar estes vinhos em duas categorias (bons/ruins)

```python
# Quando a qualidade for maior que 5 = bom; caso contrário = ruim.

df['quality_bin'] = df['quality'].apply(lambda x : "bom" if x > 5 else "ruim")
```

```python
df_bin = df.drop(columns=['quality'])
df_bin
```

```python
# Salva o dataset para uso posterior

df_bin.to_csv('winequality-red-binary.csv', index=False)
```
